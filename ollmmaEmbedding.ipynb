### Ollama Embedding

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings=OllamaEmbeddings(model="nomic-embed-text") 
embeddings

```

```python
r1 = embeddings.embed_documents([
    "Alpha is the first letter of Greel alphabet",
    "Beta is the second letter of Greek alphabet",
])
```

```python
r1
```

```python
print(len(r1[0])) # dimension size
print(r1[:2]) # check vectors
```

```python
embeddings.embed_query("What is the scond letter of Greek alphabet")
```
